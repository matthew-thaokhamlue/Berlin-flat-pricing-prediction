```python
import pandas as pd
from pycaret.regression import *
```

# 2. Load Data

```python
df = pd.read_csv('berlin-houses.csv')
```

```python
df = df.drop(columns=['id','currency','cold_price','currency','short_listed',
                                'created_date','published_date',
                                'modified_date','address','link',
                                'listing_type','localhost_date','no_longer_available',
                                'no_longer_available_date','quarter'])
```

```python
df.dtypes
```

# 3. Train and Evaluate Model

```python
cat_features = ['postcode_id']
```

```python
experiment = setup(df, target='warm_price', categorical_features=cat_features, profile=True)
```

```python
best_model = compare_models()
```

```python
# evaluate_model(best_model)
```

```python
rf = create_model('rf')
```

```python
tuned_rf = tune_model(rf)
```

```python
bagged_rf = ensemble_model(rf)
```

# 4. Test Model

```python
predict_model(bagged_rf, df.head())
```

# 5. Save Model

```python
save_model(best_model, model_name='rf-model')
```
